```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('dark_background')
# Parameters to our function
eps = 0.4
a, b, c= 3, 1, -1
# Generating the corresponding data
x = np.linspace(-1, 1, 200)
poly = np.poly1d([a, b, c])
epsilon = np.random.uniform(-eps, eps, x.shape)
y = poly(x) + epsilon
plt.scatter(x, y)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
# Our data to fit
X = x[:, np.newaxis]
# Degree of our polynomial
degree = 2;
# Create our polynomial model for regression
model = make_pipeline(PolynomialFeatures(degree), LinearRegression())
# Fit the parameters of this model
model.fit(X, y);
x_plot = np.linspace(-1, 1, 200)
plt.plot(x_plot, model.predict(x_plot[:, np.newaxis]), c='r')
plt.scatter(x, y)
print(var(model.steps[0][1]))
```

```python
true_w0 = 2
true_w1 = 3
eps = 0.1
x = np.random.rand(200)
y = true_w0 + true_w1 * x + (eps * np.random.randn(200))
plt.scatter(x, y)
plt.xlabel('year')
plt.ylabel('temperature')
```

```python
def mse_loss(y, y_bar):
    return np.sum((y - y_bar) ** 2)

n_iter = 1000
lr = 0.001
w_0, w_1 = np.random.randn(), np.random.randn()

for i in range(n_iter):
    # 1. Calculate the predictions
    y_bar = w_1 * x + w_0
    # 2. Compute the loss
    loss = mse_loss(y, y_bar)
    # 3. Calculate the gradients
    dw_1 = 2 * np.sum((y_bar - y) * x)
    dw_0 = 2 * np.sum((y_bar - y))
    # 4. Update the parameters
    w_1 -= lr * dw_1
    w_0 -= lr * dw_0

p = plt.scatter(x, y)
plt.plot(x, true_w1 * x + true_w0, linewidth=6, alpha=0.6, color="red")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
# Properties of the problem
n_observations = 200
noise = 0.2
c1_center = [-2, -1]
c2_center = [2, 1]
# Create points
x_coords, y_class = make_blobs(n_samples=n_observations, centers=[c1_center, c2_center], n_features=2, cluster_std=0.55)
x_inputs = x_coords + (noise * np.random.randn(n_observations, 2))
print(x_inputs.shape)
# Plot the corresponding pattern
plt.scatter(x_inputs[y_class == 0, 0], x_inputs[y_class == 0, 1], c='r')
plt.scatter(x_inputs[y_class == 1, 0], x_inputs[y_class == 1, 1], c='g')
```

```python
import jax
import jax.numpy as jnp
from jax import grad, jit, random
key = random.PRNGKey(42)
x_jax = jnp.asarray(x_inputs)
y_jax = jnp.asarray(y_class)
eta, n_epochs = 0.01, 50
# Perform key splitting
key_w, key_b, key = random.split(key, 3)
# Initialize the weights
weights = random.normal(key_w, (1, 2));
bias = random.normal(key_b, (1, 1));
# Single neuron full function
def loss_function(x, y, w, b):
    # Forward pass
    y_bar = jax.nn.sigmoid(jnp.dot(w, x.T) + b)
    # Error computation
    error = jnp.sum((y_bar - y) ** 2)
    return error
# Gradient of the loss
grad_loss_function = jit(grad(loss_function, argnums=[2, 3]))
# Update loop
for i in range(n_epochs):
    # 1. Forward pass and gradients
    grads = grad_loss_function(x_jax, y_jax, weights, bias)
    # 3. Update weights and bias
    weights -= eta * grads[0]
    bias -= eta * grads[1]
```

```python
plt.figure()
plt.scatter(x_jax[:, 0], x_jax[:, 1])
x_classif = np.linspace(-3, 3, 100)
y_class = (-np.dot(weights[0, 0], x_classif) - bias[0]) / weights[0, 1]
plt.plot(x_classif, y_class, color='r')
plt.ylim(-5, 5)
```
