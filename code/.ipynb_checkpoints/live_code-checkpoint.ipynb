```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('dark_background')
a, b, c = 3, 1, -1
# Base (X) data
x = np.linspace(-1, 1, 200)
# Our target function
poly = np.poly1d([a, b, c])
# Adding observation noise
eps = 1
noise = np.random.uniform(-eps, eps, x.shape)
y = poly(x) + noise
plt.scatter(x, y)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
# Constructing our model
p_degree = 2
model = make_pipeline(PolynomialFeatures(p_degree), LinearRegression())
```

```python
# Adding one dimension
X = x[:, np.newaxis]
model.fit(X, y);
```

```python
x_predict = np.linspace(-1.5, 3, 100)
y_predict = model.predict(x_predict[:, np.newaxis])
plt.figure()
plt.scatter(x, y)
plt.plot(x_predict, y_predict, c='r')
```

```python
# Classify some data 
import librosa as li
cat_signal, sr = li.load('01_cat.wav')
```

```python
import tensorflow_hub as hub
model = hub.load('https://tfhub.dev/google/yamnet/1')
# Create our silence chunk
sr = 16000
len = 3
silence = np.zeros(3 * 16000, dtype=np.float32)
```

```python
scores, embeddings, spectro = model(silence)
```

```python
import tensorflow as tf
import io
import csv
# Find the name of the class with the top score when mean-aggregated across frames.
def class_names_from_csv(class_map_csv_text):
  """Returns list of class names corresponding to score vector."""
  class_map_csv = io.StringIO(class_map_csv_text)
  class_names = [display_name for (class_index, mid, display_name) in csv.reader(class_map_csv)]
  class_names = class_names[1:]  # Skip CSV header
  return class_names
class_map_path = model.class_map_path().numpy()
# Class names
class_names = class_names_from_csv(tf.io.read_file(class_map_path).numpy().decode('utf-8'))
```

```python
print(class_names[scores.numpy().mean(axis=0).argmax()])
```
