# OSC / MaxMSP connection

```python
from pythonosc import osc_server
from pythonosc import dispatcher
from pythonosc import udp_client
# Address of our server
addr = "localhost"
addr = "127.0.0.1"
port_server = 1243
port_client = 1239

client = udp_client.SimpleUDPClient(addr, port_client)
client.send_message("/synth/frequency", 440.0)

def volume_func(address, volume):
    # LAUNCH DEEP COMPUTATION !
    client.send_message("/synth/volume", 2 * volume) 
    print(f"Received volume : {volume}")

# Message to function mapping
dispatch = dispatcher.Dispatcher()
# Create the mapping
dispatch.map("/synth/volume", volume_func)
# Server object
server = osc_server.BlockingOSCUDPServer((addr, port_server), dispatch)
server.serve_forever()
```
